```python
filename='../googlenet_old/fig/apple1.png'

img = cv2.imread(filename)

if img is None:
    print("이미지 없음")
    sys.exit()

## dnn Model
model='./bvlc_googlenet.caffemodel' # 네트워크의 구조가 들어가있음(학습된 파라메타임)
config = '../googlenet_old/deploy.prototxt'# 위 모델에 대한 설명이 들어가있음(구조가 어떠하다)

net = cv2.dnn.readNet(model, config)

if net.empty():
    print('network load failed')
    sys.exit()

########## 클래스 이름 불러오기

classNames = []
with open('./classification_classes_ILSVRC2012.txt', 'rt') as f: # 천개의 오브젝트
    classNames = f.read().rstrip('\n').split('\n') #마지막 \n을 지우고 스플릿 하라
    
# print(classNames)
blob = cv2.dnn.blobFromImage(img, 1, (224,224), (104, 117, 123), swapRB = False) #우리의 이미지를 구글넷에 맞게 변경해 줌
net.setInput(blob)
prob = net.forward() # 학습한 모델을 가지고 확률로 변경

print(prob.shape)

out=prob.flatten()
print(out.shape) # 사용하기 편하기 위해 축소함

classid = np.argmax(out)
print(classid)

confience = out[classid]
print(confience)

print(classNames[classid])

text = f'{classNames[classid]} ({confience*100:3.2f})%' # 텍스트를 쓰는 형식(f를 쓰는 이유는 변수 값을 가져다 쓰겠다!)

cv2.putText(img, text, (10,30), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,225), 1, cv2.LINE_AA)

cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```
