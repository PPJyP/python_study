```python
#데이터프레임의 행과 열을 바꿈
import pandas as pd
```

```python
exam_data={'이름':['서준','우현','인아'],
           '수학':[90,80,70],
           '영어':[98,89,95]}
df =pd.DataFrame(exam_data)
print(df)

df.set_index('이름',inplace=True) #이름을 기준으로 이덱스 변경(shift+Tab으로 확인)
print(df)
```

```python
df=df.T
print(df)
```

```python
#reindex(새로운 인덱스 배열):기존의 자료변경 안함, 새로운 데이터프레임 반환
dict_data={'c0':[1,2,3],'c1':[1,2,3],'c2':[1,2,3],'c3':[1,2,3]}
df=pd.DataFrame(dict_data,index=['r1','r2','r3'])
df
```

```python
#index를['r0','r1','r2','r3','r4']
new_index=['r0','r1','r2','r3','r4']
ndf=df.reindex(new_index)
print(ndf)

#존재하지 않는 인덱스의 값을 0으로 채워줌
ndf1=df.reindex(new_index,fill_value=0)
ndf1

#뒤에 있는 행을 따름('bfill') / 앞에 있는 행을 따름('ffill')
ndf1=df.reindex(new_index,method='bfill')
ndf1
```

```python
#reset_index():인덱스 초기화
print(df)
print(df.reset_index()) #기존의 인덱스가 컬럼으로 변환, 인덱스는 초기화
print(df.reset_index(drop=True))# 기존의 인덱스를 삭제, 인덱스는 초기화
```

```python
#인덱스를 기준으로 정렬
df.sort_index(ascending=False)
```

```python
#컬럼을 기준으로 정렬
df.loc[0,'c0']=2
df.loc[0,'c2']=4
df.sort_values(by=['c0','c2'],ascending=False) #c0기준으로 내림차순 정리하다가 동일한 값은 c2 기준으로 내림차순 정렬
```

```python
#인덱스를 설정 df.index=[]
#컬럼명을 설정 df.columns=[]
#인덱스의 이름을 변경 df.rename({기존인덱스:새로운인덱스,..},inplace=True)
#컬럼명을 변경 df.rename({기존컬럼:새로운컬럼},inplace=True)
#인덱스를 삭제 df.drop([인덱스명], axis=0)
#컬럼 삭제 df.drop([컬럼명],axis=1)
#새로운 인덱스 이름 부여: ndf=df.reindex(new_index,[fill_value=0 또는 method='f/bfill'])
#인덱스 제거 df.reset_index([drop=True]) 인덱스를 컬럼으로 변환하지 않음
#인덱스 기준으로 정렬 df.sort_index([asending=])
#컬럼 기준으로 정렬 df.sort_values(by=[컬럼명])
```

```python
#시리즈 연산
student=pd.Series({'국어':100,'수학':90,'영어':90})
print(student)

print(student-10)
print(student*2)
```

```python
# 가나다 순으로 정렬해서 보여줌
student1=pd.Series({'국어':100,'수학':90,'영어':90})
student2=pd.Series({'수학':60,'영어':100,'국어':70,'과학':100}) #none값을 더하면 무조건 NaN

print(student1+student2)
```

```python
#시리즈 연산 메서드 : 시리즈.add(시리즈1, fill_value=0)
print(student1.add(student2,fill_value=0),'\n')
print(student1.sub(student2,fill_value=0),'\n')
print(student1.mul(student2,fill_value=0),'\n')
print(student1.div(student2,fill_value=0),'\n')
```
